```python
import sys
import os
notebook_dir = os.getcwd()
folder_p_path = os.path.join(notebook_dir, '..')
sys.path.append(os.path.abspath(folder_p_path))
from BETpy.bet import DDMbet
import numpy as np

ddmbet = DDMbet()
dataX = [x for x in range(100)]
dataY = np.zeros(100)
for i in range(100):
    if i<25:
        dataY[i]=50+i*2
    if i>=25 and i<50:
        dataY[i] = (i-25)*2
    if i>=50 and i<75:
        dataY[i] = 50+(i-50)*2+1
    if i>=75:
        dataY[i] = (i-75)*2+1

min_p, mincount, min_cross= ddmbet.runBET(dataX, dataY, plot=True)

print(f"Minimum P-value: {min_p}")
print(f"Group 0 count: {mincount}")
print(f"Binary Interaction: {min_cross}")
```

```python
import pandas as pd
import numpy as np
import math
np.random.seed(0)
mydf=pd.DataFrame()
ddmbet = DDMbet()
dataX = np.random.uniform(0, 20*math.pi, size=200)
mydf[f"col_0"] =dataX
for i in range(1,100):
    dataY =[math.sin(0.3*i*x) for x in dataX]
    mydf[f"col_{i}"] =dataY

xvar = mydf['col_2']
yvar = mydf['col_3']
min_p, mincount, min_cross= ddmbet.runBET(xvar, yvar, max_depth=10, plot=True)
print(f"Minimum P-value: {min_p}")
print(f"Group 0 count: {mincount}")
print(f"Binary Interaction: {min_cross}")
```
